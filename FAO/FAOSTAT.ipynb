```python
def wait_until_done(path_to_file:str,max_wait:int,step_wait:int,stop_count:int):
    """
    checks the file size of "path_to_file every" every "step_wait" seconds until the file size stays the same for 
    "stop_count" number of retries. Unless "max_wait" seconds is reached then it stops everything
    """

    start = time.time()
    size = 0
    same_count = 0
    while True:
        new_size = os.path.getsize(path_to_file)
        if new_size != size:
            same_count = 0
            size = new_size
            time.sleep(step_wait)
        else:
            same_count += 1
        if same_count == stop_count:
            print('Download complete')
            break
            
        if time.time() > start + max_wait:
            print('max_wait reached')
            break
    return 

wait_until_done(r"download/path/to/file.xlsx",30,1,3)
```

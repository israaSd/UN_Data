```python
column = ['country']
content = driver.find_element(By.CLASS_NAME,'o-country-score-card')
rankings_overview = content.find_element(By.CLASS_NAME,'o-performance-overview').find_elements(By.CLASS_NAME,'o-pillar')
for ranking_overview in rankings_overview:
    values_overview = ranking_overview.text.split('\n')
    col_o_score = f'Score_{values_overview[0]}'
    col_o_rank = f'Rank_{values_overview[0]}'
    column.append(col_o_score)
    column.append(col_o_rank)
rankings_indicators = content.find_element(By.CLASS_NAME,'o-indicators-table').find_elements(By.CLASS_NAME,'m-row')[1:]
for ranking_indicators in rankings_indicators:
    values_indicators = ranking_indicators.text.split('\n')
    col_i_rank = f'Rank_{values_indicators[0]}'
    col_i_score = f'Score_{values_indicators[0]}'
    col_i_male = f'Male_{values_indicators[0]}'
    col_i_female = f'Female_{values_indicators[0]}'
    column.append(col_i_rank)
    column.append(col_i_score)
    column.append(col_i_male)
    column.append(col_i_female)
    
country = driver.find_element(By.CLASS_NAME,'a-level-2-label-container.a-header-value')
country.click()
time.sleep(1)
countries = driver.find_element(By.CLASS_NAME,'m-level-2-select-container')
driver.find_element(By.CLASS_NAME,'css-o9lev6-control').click()
countries_option = countries.find_elements(By.CLASS_NAME,'css-mugoag-option')
country.click()
row = []
for c in range(len(countries_option)+1):
    time.sleep(2)
    l = []
    country_name = driver.find_element(By.CLASS_NAME,'a-level-2-label-container.a-header-value').text
    l.append(country_name)
    content = driver.find_element(By.CLASS_NAME,'o-country-score-card')
    rankings_overview = content.find_element(By.CLASS_NAME,'o-performance-overview').find_elements(By.CLASS_NAME,'o-pillar')
    for ranking_overview in rankings_overview:
        values_overview = ranking_overview.text.split('\n')
        for n in range(1,3):
            value_overview = values_overview[n]
            l.append(value_overview)
    rankings_indicators = content.find_element(By.CLASS_NAME,'o-indicators-table').find_elements(By.CLASS_NAME,'m-row')[1:]
    for ranking_indicators in rankings_indicators:
        values_indicators = ranking_indicators.text.split('\n')
        import itertools as it
        for m in it.chain(range(2, 4), range(7, 9)):
            l.append(values_indicators[m])
    row.append(l)
    try:
        country = driver.find_element(By.CLASS_NAME,'a-level-2-label-container.a-header-value')
        country.click()
        time.sleep(1)
        countries = driver.find_element(By.CLASS_NAME,'m-level-2-select-container')
        driver.find_element(By.CLASS_NAME,'css-o9lev6-control').click()
        time.sleep(1)
        countries_option = countries.find_elements(By.CLASS_NAME,'css-mugoag-option')
        countries_option[c].click()
        time.sleep(2)
    except:
        pass
df = pd.DataFrame(row,columns=column) 
df.to_excel(f'{base_path}\\WORLD ECONOMIC FORUM 2023.xlsx',index = False)
```

```python
url = 'https://www.gulftalent.com/api/jobs/search?condensed=false&config%5Bfilters%5D=ENABLED&config%5Bresults%5D=UNFILTERED&filters%5Bcountry%5D%5B0%5D=10111111000000&filters%5Bcountry%5D%5B10%5D=10229118000000&filters%5Bcountry%5D%5B1%5D=10111112000000&filters%5Bcountry%5D%5B2%5D=10111114000000&filters%5Bcountry%5D%5B3%5D=10111113000000&filters%5Bcountry%5D%5B4%5D=10229362000000&filters%5Bcountry%5D%5B5%5D=10229117000000&filters%5Bcountry%5D%5B6%5D=10111116000000&filters%5Bcountry%5D%5B7%5D=10111115000000&filters%5Bcountry%5D%5B8%5D=10229411000000&filters%5Bcountry%5D%5B9%5D=10229120000000&include_scraped=1&limit=10000&search_keyword=&version=2'
response = requests.get(url)
dataflows = response.json()
dataflows
#dataflows['results']['data']
```

```python
company_name = []
title = []
location = []
category_name = []
company_url =[]
industry_name = []
codes_dict = []
link = []
for i in range(len(dataflows['results']['data'])):
    code = {key: dataflows['results']['data'][i][key] for key in dataflows['results']['data'][i].keys()
              & {'title', 'category_name', 'company_name', 'company_url','industry_name','location','link'}}
    #codes_dict.append(code)
    #print(code)
    company_name.append(code['company_name'])
    title.append(code['title'])
    location.append(code['location'])
    category_name.append(code['category_name'])
    company_url.append(code['company_url'])
    industry_name.append(code['industry_name'])
    link.append(code['link'])
code
```
